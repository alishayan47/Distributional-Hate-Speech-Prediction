```python
from datasets import load_dataset
import pandas as pd
```

Loading the dataset from Hugging Face.

```python
dataset_name = "ucberkeley-dlab/measuring-hate-speech"
dataset = load_dataset(dataset_name)
```

Converting the dataset to a pandas DataFrame for easier manipulation and calculating the average number of annotations per comment ID.

```python
df = dataset['train'].to_pandas()

# Calculate the number of annotations per comment_id
annotations_per_comment = df.groupby('comment_id').size().reset_index(name='num_annotations')

# Calculate the average number of annotations
average_annotations = annotations_per_comment['num_annotations'].mean()

print(f"Average number of annotations per comment ID: {average_annotations:.2f}")

print("\nExamples of comment IDs and their annotation counts:")
display(annotations_per_comment.head())
```

Displaying some rows from the original dataset to show what the annotations look like for a few comment IDs.
