```python
import os
import pandas as pd
```

```python
mhs = pd.read_csv("../data/processed/mhs_annotations_clean.csv")

print("Loaded shape:", mhs.shape)
```

```python
# Target columns

women_target_col = "target_gender_women"

immigrant_target_cols = [
    "target_origin_immigrant",
    "target_origin_migrant_worker",
    "target_origin_specific_country",
    "target_origin_undocumented"
]
```

```python
# Converting target columns to numeric just in case they loaded as True/False strings

target_cols = [women_target_col] + immigrant_target_cols

for col in target_cols:
    mhs[col] = mhs[col].astype(int)
```

```python
# Women-targeted subset

women_df = mhs[mhs[women_target_col] == 1].copy()

print("Women-targeted annotation rows:", women_df.shape[0])
print("Unique women-targeted comments:", women_df["comment_id"].nunique())
```

```python
# Immigrant/origin-targeted subset

immigrant_df = mhs[mhs[immigrant_target_cols].eq(1).any(axis=1)].copy()

print("Immigrant-targeted annotation rows:", immigrant_df.shape[0])
print("Unique immigrant-targeted comments:", immigrant_df["comment_id"].nunique())
```

```python
# Quick overlap check

overlap_comments = set(women_df["comment_id"]).intersection(set(immigrant_df["comment_id"]))

print("Comments appearing in both subsets:", len(overlap_comments))
```

```python
# Save filtered datasets

os.makedirs("../data/processed", exist_ok=True)

women_path = "../data/processed/mhs_women_targeted_annotations.csv"
immigrant_path = "../data/processed/mhs_immigrant_targeted_annotations.csv"

women_df.to_csv(women_path, index=False)
immigrant_df.to_csv(immigrant_path, index=False)

print("Saved:", women_path)
print("Saved:", immigrant_path)
```
